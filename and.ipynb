```python
# ham sigmoid chuan hoa du lieu
def sigmoid(x):
    return 1. / (1. + np.exp(-x))
```

```python
# DAO HAM SIGMOID
def sigmoid_derivative(x):
    return sigmoid(x) * (1. - sigmoid(x))
```

```python
for i in range(iterations):
    out = np.zeros((4,1))
    for j in range(numInput):
#         y = bias * weights[0,0] + input[j, 0] * weights[1,0] + input[j, 1] * weights[2,0]
        z1 = bias + input[j, 0] * weights[0, 0] + input[j, 1] * weights[1, 0]
        a1 = sigmoid(z1)
        out[j] = a1
        error = a1 - expect_out[j]
        
        # backpropagation
        delta_bias = sigmoid_derivative(z1)
        delta_weights00 = sigmoid_derivative(z1) * input[j, 0]
        delta_weights01 = sigmoid_derivative(z1) * input[j, 1]
        
        bias -= learning_rate * error * delta_bias
        weights[0, 0] -= learning_rate * error * delta_weights00
        weights[1, 0] -= learning_rate * error * delta_weights01
#         print ("loss: %s", error)
    print (out)
```
