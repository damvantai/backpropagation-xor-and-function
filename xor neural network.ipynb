```python
import numpy as np
```

```python
input = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
input
```

```python
numInput = 4
```

```python
expect_out = np.array([[0], [1], [1], [0]])
```

```python
expect_out
```

```python
weight = np.random.randn(2, 2)
```

```python
weight
```

```python
bias = np.random.randn(2, 1)
```

```python
theta = np.random.randn(2, 1)
```

```python
c = np.random.randn(1, 1)
```

```python
learning_rate = 0.7
```

```python
# ham sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))
```

```python
iterations = 6000
```

```python
for i in range(iterations):
    out = np.zeros((4, 1))
    for j in range(numInput):
        # feed forward
        z1 = input[j, 0] * weight[0, 0] + input[j, 1] * weight[0, 1] + bias[0,0]
        z2 = input[j, 0] * weight[1, 0] + input[j, 1] * weight[1, 1] + bias[1,0]
        a1 = sigmoid(z1)
        a2 = sigmoid(z2)
        z3 = theta[0, 0] * a1 + theta[1, 0] * a2 + c
        a3 = sigmoid(z3)
        
        out[j] = a3
        
        # error 
        error = (a3 - expect_out[j])
        
        # backpropagation
        delta_theta00 = sigmoid_derivative(z3) * a1
        delta_theta10 = sigmoid_derivative(z3) * a2
        delta_c = sigmoid_derivative(z3)
        delta_b00 = sigmoid_derivative(z3) * theta[0, 0] * sigmoid_derivative(z1) 
        delta_b10 = sigmoid_derivative(z3) * theta[1, 0] * sigmoid_derivative(z2)
        delta_w00 = sigmoid_derivative(z3) * theta[0, 0] * sigmoid_derivative(z1) * input[j, 0]
        delta_w01 = sigmoid_derivative(z3) * theta[0, 0] * sigmoid_derivative(z1) * input[j, 1]
        delta_w10 = sigmoid_derivative(z3) * theta[1, 0] * sigmoid_derivative(z2) * input[j, 0]
        delta_w11 = sigmoid_derivative(z3) * theta[1, 0] * sigmoid_derivative(z2) * input[j, 1]
        
        # update parameter
        theta[0, 0] -= error * learning_rate * delta_theta00
        theta[1, 0] -= error * learning_rate * delta_theta10
        c -= error * learning_rate * delta_c
        bias[0, 0] -= error * learning_rate * delta_b00
        bias[1, 0] -= error * learning_rate * delta_b10
        weight[0, 0] -= error * learning_rate * delta_w00
        weight[0, 1] -= error * learning_rate * delta_w01
        weight[1, 0] -= error * learning_rate * delta_w10
        weight[1, 1] -= error * learning_rate * delta_w11
        
        
    print (out)
print ("weight[0, 0]: %s", weight[0, 0])
print ("weight[0, 1]: %s", weight[0, 1])
print ("weight[1, 0]: %s", weight[1, 0])
print ("weight[1, 1]: %s", weight[1, 1])
print ("bias[0, 0]: %s", bias[0, 0])
print ("bias[1, 0]: %s", bias[1, 0])
```

```python
def feed_forward(x1, x2):
    z1 = x1 * weight[0, 0] + x2 * weight[0, 1] + bias[0,0]
    z2 = x1 * weight[1, 0] + x2 * weight[1, 1] + bias[1,0]
    a1 = sigmoid(z1)
    a2 = sigmoid(z2)
    z3 = theta[0, 0] * a1 + theta[1, 0] * a2 + c
    a3 = sigmoid(z3)
    print (a3)
```
